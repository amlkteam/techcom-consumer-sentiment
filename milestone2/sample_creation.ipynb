```python
from random import randrange
import pandas as pd

df = pd.read_csv('data/twitter_french_results_last.csv')

sample = {}

for i in range(20):
    rand = randrange(len(df))
    sample[rand] = df.iloc[rand]
    
sample_df = pd.DataFrame.from_dict(sample).transpose()
sample_df
```

```python
sample_df.to_csv('sample_df.csv')
```
