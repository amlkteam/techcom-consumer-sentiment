```python
import pandas as pd
```

```python
header = ['ind','created_at','member_id','text','textLength','source','user','verified','screen_name','verified_type','user_description','bid','longText']
```

```python
applecn_df=pd.read_csv("weibo_%23%E8%8B%B9%E6%9E%9C%23_scrap_fullstr.csv",encoding = "gb18030",names=header,index_col=0)
```

```python
applecn_df.head()
```

```python
appleen_df=pd.read_csv("weibo_Apple_scrap_fullstr.csv",encoding = "gb18030",index_col=0)
```

```python
appleen_df.head()
```

```python
applecn_filtered = applecn_df.drop(columns=["verified","source","longText"])
applecn_filtered.head()
```

```python
appleen_df_filtered=appleen_df.drop(columns=["verified","source","longText"])
appleen_df_filtered.head()
```

```python
combined_df=pd.concat([applecn_filtered,appleen_df_filtered], sort = False)
combined_df =combined_df.reset_index()
combined_df = combined_df.drop(columns="index")
combined_df.head()
```

```python
comb_df =combined_df.drop_duplicates() #734 rows after dropping duplicates
```

```python
comb_df.query("textLength == 'textLength'")
```

```python
comb_df_cleaned=comb_df.drop([559,581,612])
```

```python
comb_df_cleaned=comb_df_cleaned.fillna(0)
```

```python
comb_df_cleaned['textLength'].unique()
```

```python
comb_df_cleaned['textLength']=comb_df_cleaned['textLength'].astype(int)
```

```python
sum(comb_df_cleaned['textLength'].astype(int))
```

```python
comb_df_cleaned['verified_type'].astype(int)
```

```python
comb_df_cleaned['verified_type'] = comb_df_cleaned['verified_type'].astype(int)
```

```python
#669 rows
comb_person = comb_df_cleaned.query("verified_type == -1 or verified_type == 0 or verified_type == 200 or verified_type == 220")
```

```python
comb_person.head()
```

```python
sum(comb_person['textLength'])
```

```python
comb_person['text']
```

```python
comb_nobid =comb_person[['created_at','member_id','screen_name','text','textLength','verified_type','user_description','user']] #669 rows
```

```python
len(comb_nobid['text'].unique())  #only 568 rows after removing duplicate texts
```

```python
unique_ind=comb_nobid['text'].drop_duplicates().index
```

```python
unique_Apple_comments=comb_nobid.loc[unique_ind]
```

```python
unique_Apple_comments.loc[727]
unique_Apple_comments=unique_Apple_comments.drop(727)
```

```python
unique_Apple_comments['created_at']=unique_Apple_comments['created_at'].str.extract("(\d{4}-\d{2}-\d{2})")
```

```python
sum(unique_Apple_comments['textLength']) # after all cleaning still has 13,5000 words
```

```python
unique_Apple_comments.to_csv("Apple_comb_personsonly.csv", index=True, encoding='gb18030',header=True)
```
