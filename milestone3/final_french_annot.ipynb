```python
final_df = molly.copy(deep = True)
final_df.head()
```

```python
final_df['rating'] = 'blank'
final_df.head()
```

```python
n = final_df.columns.get_loc("rating")
n
```

```python
for i in range(len(molly)):
    if molly.iloc[i, 7] == joanna.iloc[i, 7]:
        final_df.iloc[i, 7] = molly.iloc[i, 7]
    elif molly.iloc[i, 7] != joanna.iloc[i, 7]:
#         print(i)
#         print(molly.iloc[i, 7])
#         print(joanna.iloc[i, 7])
        rat_m = molly.iloc[i, 7]
        rat_j = joanna.iloc[i, 7]
        if rat_m == "Irrelevant" or rat_j == "Irrelevant":
            final = "Irrelevant"
        elif rat_m == "Neutral" or rat_j == "Neutral":
            final = "Neutral"
        elif rat_m.startswith("Very") or rat_j.startswith("Very"):
            if rat_m[5] == "N" or rat_j[5] == "N":
                final = "Negative"
            if rat_m[5] == "P" or rat_j[5] == "P":
                final = "Positive"
        elif rat_m == "Negative" and rat_j == "Positive" or rat_m == "Positive" and rat_j == "Negative":
            final = "Neutral"
        final_df.iloc[i, 7] = final
```
